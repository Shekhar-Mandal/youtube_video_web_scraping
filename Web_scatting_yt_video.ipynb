```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time



driver = webdriver.Chrome()

driver.get(f"https://www.youtube.com/watch?v=nILY9FSj-nU/")
```

```python
import numpy as np


# /html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/div[11]/ytd-watch-next-secondary-results-renderer/div[2]/ytd-compact-video-renderer[1]/div[1]/div/div[1]/a/h3/span
# /html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/div[11]/ytd-watch-next-secondary-results-renderer/div[2]/ytd-compact-video-renderer[2]/div[1]/div/div[1]/a/h3/span
title_of_video=[]
views=[]
durations=[]


for i in range(1,30):
    try:
        title=driver.find_element(By.XPATH,f'/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/div[11]/ytd-watch-next-secondary-results-renderer/div[2]/ytd-compact-video-renderer[{i}]/div[1]/div/div[1]/a/h3/span').text
        title_of_video.append(title)
        print(title)
    except:
        title_of_video.append(np.nan)
    try:
        views_of_video=driver.find_element(By.XPATH,f'/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/div[11]/ytd-watch-next-secondary-results-renderer/div[2]/ytd-compact-video-renderer[{i}]/div[1]/div/div[1]/a/div/ytd-video-meta-block/div[1]/div[2]/span[1]').text
        views.append(views_of_video)
        print(views_of_video)
    except:
        views.append(np.nan)  
    try:
        duration_of_video=driver.find_element(By.XPATH,f'/html/body/ytd-app/div[1]/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[2]/div[11]/ytd-watch-next-secondary-results-renderer/div[2]/ytd-compact-video-renderer[{i}]/div[1]/div/div[1]/a/div/ytd-video-meta-block/div[1]/div[2]/span[2]').text
        durations.append(duration_of_video)
        print(duration_of_video)
    except:
        duration.append(np.nan)
```

# conbvert the lists intpo the dataframe 

```python
#!pip install pandas 

import pandas as pd

dicts={'title':title_of_video,'views':views,'durations':durations}
dicts
```

```python
df=pd.DataFrame(dicts)
```
